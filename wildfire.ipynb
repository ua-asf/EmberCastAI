```python
import zipfile
import os

kmz_file = "/home/dmmaltos/Desktop/wildfire prediction/data/20240919_BachelorCx_IR(9).kmz"  # Ensure this is a complete path with the .kmz extension
output_folder = "/home/dmmaltos/Desktop/wildfire prediction/extracted/"  # Folder to save extracted images

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

try:
    # Open the KMZ file
    with zipfile.ZipFile(kmz_file, 'r') as kmz:
        # List all files to verify contents
        print("Files in KMZ:", kmz.namelist())
        
        # Find all PNG files
        png_files = [name for name in kmz.namelist() if name.endswith('.png')]
        if not png_files:
            print("No .png files found in the KMZ archive.")
        else:
            # Extract each PNG file
            for png_filename in png_files:
                with kmz.open(png_filename) as png_file:
                    # Define the output path for each image
                    output_path = os.path.join(output_folder, os.path.basename(png_filename))
                    with open(output_path, 'wb') as out_file:
                        out_file.write(png_file.read())
                    print(f"Extracted {png_filename} to {output_path}")

except zipfile.BadZipFile:
    print("Error: The file is not a valid KMZ file.")
except Exception as e:
    print("An error occurred:", e)
```
